```python
import matplotlib.pyplot as plt

# Sample data
x = ["davinci-002",
     "davinci-003",
     "gpt-3.5-instruct",
#      "gpt-3.5-turbo",
     "gpt-4",
     "llama-3-70b"]
y1 = [0.477, 0.53, 0.543, 0.550, 0.603]
y2 = [0.213, 0.281, 0.19, 0.263, 0.335]

fig, ax1 = plt.subplots(figsize=(22, 14))

# Plot data for the first y-axis, in two parts to exclude the line between the 4th and 5th points
ax1.plot(x[:4], y1[:4], 'g-', marker='o', markersize=40, linewidth=10, label="Acc")  # First part
ax1.scatter(x[4], y1[4], color='g', marker='o', s=1600, label="Acc")  # Second part
ax1.set_ylabel('Accuracy', color='g', fontsize=50)
ax1.tick_params('y', colors='g')
ax1.set_ylim(0.45, 0.65)

# Create a second y-axis sharing the same x-axis
ax2 = ax1.twinx()
ax2.plot(x[:4], y2[:4], 'b-', marker='o', markersize=40,linewidth=10,  label='ECE')  # First part
ax2.scatter(x[4], y2[4], color='b', marker='o', s=1600, label='ECE')  # Second part
ax2.set_ylabel('ECE Score', color='b', fontsize=50)
ax2.tick_params('y', colors='b')
ax2.set_ylim(0, 2)

# Add a dashed vertical line between the 4th and 5th dots
ax1.axvline(x=3.5, color='k', linestyle='--')

# loop over all axes

ax1.tick_params(axis='y', which='major', labelsize=30, pad=40)
ax2.tick_params(axis='y', which='major', labelsize=30, pad=40)
ax1.tick_params(axis='x', which='major', labelsize=40, pad=40)

ax1.legend(loc='upper left', fontsize=40)
ax2.legend(loc='upper right', fontsize=40)

# fig.tight_layout()
plt.show()
```
